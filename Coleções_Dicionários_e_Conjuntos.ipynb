- COLEÇÕES

 - Dicionários

  Os dicionários são representadas com { } e para atribuir um parâmetro com os índices utilizamos o [ ]. Podemos também percorrer um dicionário com o comando FOR. O índice vai caminhar e printar todos os valores armazenados naquele espaço de memória.

  com o ***del(coleta)['nome']*** podemos apagar somente um espaço da memória disposta no dicionário. Para apagar o dicionário inteiro podemos usar o ***del(coleta)***

```python
coleta = {'Will':1, 'Victoria':2, 'Ana C':3, 'Ana P':4}
print('Índice da coleta:',coleta['Will'])
print('Índice da coleta:',coleta['Victoria'])
print('---')
#del(coleta)['Ana P']
```

- Fuções (métodos):

 - ***Items()*** -> Mostra o dicionário no formato ***('chave',valor)***
 - ***keys()*** -> Mostra apenas as chaves (nomes)
 - ***values()*** -> Mostra apenas os valores (índices)
 - Se apenas printarmos ela sai norma com {'Chave':valor}
 - Com o ***update()*** atualizamos e adicionamos chaves e valores novos para o dicionário.

```python
coleta = {'Will':1, 'Victoria':2, 'Ana C':3, 'Ana P':4}
print(coleta.items())
print('---')
print(coleta.keys())
print('---')
print(coleta.values())
print('---')

coleta2 = {'João':5, 'José':6, 'Pedro':7, 'Melissa':8}

coleta.update(coleta2)
print(coleta)
print('---')

for chave, valor in coleta.items():
    print(f'Nome: {chave}, índice: {valor}')
```

- CONJUNTOS

Formando tuplas e utilizando a função ***set ()*** para chamar o conjunto apenas, sem repetir os que estiverem repetidos.

 - A função ***intersection ()*** pega os valores atribuídos, verifica o que há de igual neles e mostra esses valores que se repetem.
 - A função ***difference ()***  pega valores atribuídos, verifica o que há de diferente neles e mostra esses valores.


```python
biomoleculas = ('proteína', 'ácido nucleico', 'carboidrato', 'lipídeo', 'carboidrato', 'lipídeo')

print('O tipo de variável é',type(biomoleculas))
print('---')
print('Tupla:',biomoleculas)
print('---')
print('Conjunto dos valores da tupla sem se repetir:',set(biomoleculas))
print('---')

c1 = {1,2,3,4,5}
c2 = {3,4,5,6,7}
c3 = c1.intersection(c2)
print('O que temos de igual entre c1 e o c2 é:',c3)
print('---')
c4 = c1.difference(c2)
print('O que tem no c1 e não tem no c2 é:',c4)
print('---')
c5 = c2.difference(c1)
print('O que tem no c2 e não tem no c1 é:',c5)
print('---')
```
