- PYTHON:

Em python não definimos o tipo da variável. A linguagem por si só analisa e escolhe de acordo com o que digitamos, salvo caso quesermos forçar um tipo específico.

```
palavra1 = 'python'
palavra2 = 'linguagem de programação'
print ('Esta', palavra2, 'se chama', palavra1)
out -> 'Esta linguagem de programação se chama python'
```

Quando queremos fazer operações com uma variável, é aconselhável que coloquemos o tipo para que no processo que fizermos, por exemplo, de adição de uma variável pela outra, não ocasione erros de tipo.

```
n1 = int(input('Informe 1ª nota: ')
n2 = int(input('Informe 2ª nota: ')

m = (n1 + n2) / 2

print (f'A média foi: {m}')
out -> 'A média foi 10'
```
