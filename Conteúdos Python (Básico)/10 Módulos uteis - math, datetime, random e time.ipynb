Bibliotecas e funções que existem em python

- Biblioteca math: https://docs.python.org/3/library/math.html
 - Funções matemáticas

```python
import math

print('Raíz quadrada de 9:', math.sqrt(9))
print('Seno de 45:', math.sin(45))
print('Cosseno de 45:', math.cos(45))
print('1000 na base de 10:', math.log(1000, 10)) # 10 x 10 x 10
print('32 na base de 2:', math.log(32, 2)) # 2 x 2 x 2 x 2 x 2
print('1000 na base do número de euler:', math.log(1000))
print('Número de euler:', math.e)
print('---')
dir(math)
```

- Biblioteca datetime: https://docs.python.org/3/library/datetime.html
 - Data e hora

```python
import datetime

print(datetime.date.today())
print(datetime.datetime.today())
print(datetime.date(2020, 7, 8))
print(datetime.datetime(2020, 5, 12, 15, 57, 29, 736584))
print('---')

data = datetime.date(2020, 1, 10)
print(f'A data informada foi {data}')
print('O dia da data é:',data.day)
print('O dia da data é:',data.month)
print('O dia da data é:',data.year)
print('---')

horario = datetime.datetime(2020, 7, 1, 15, 30, 29, 736584)
print(horario)
print('A hora informada foi:',horario.hour)
print('O minuto informado foi:',horario.minute)
print('O segundo informado foi:',horario.second)

print('---')
dir(datetime) #recursos
```

- Biblioteca random: https://docs.python.org/3/library/random.html
 - Gerar números aleatórios


```python
import random

print('Número randomico entre 0 e 1:',random.random()) #entre 0 e 1
print('Número randomico entre 1 e 5:',random.randint(1, 5)) # entre 1 e 5
print('Número randomico entre 0 e 10, contando de 2 em 2:',random.randrange(0, 10, 2)) # entre 0 e 10 em incrementos de 2, para gerar somente números pares
print('Número randomico entre 0 e 10, contando de 3 em 3:',random.randrange(0, 10, 3)) # a mesma coisa para gerar números ímpares
print('---')

x = ['K','d', 13, '34-j', 'x']
random.choice(x) # sortear um elemento de dentro de uma lista
```

- Biblioteca time: https://docs.python.org/3/library/time.html
 - Trabalhar com horas ou medirmos quanto tempo que um determinado processo levou.

```python
import time as tm # apelido, em vez de chamar time chama o tm

print(tm.time())
print('---')

antes = tm.time()
lista = []

for i in range(0, 10000):
  lista.append(i)
depois = tm.time()

intervalo = depois - antes
print(f'Tempo: {intervalo} segundos')
print('---')
print('Finalizando...')
tm.sleep(2) # após o print acima vai contar 2 segundos
print('...') # vai printar ...
tm.sleep(3) # aguarda 3 segundos
print('Até a próxima!') # finaliza
```
