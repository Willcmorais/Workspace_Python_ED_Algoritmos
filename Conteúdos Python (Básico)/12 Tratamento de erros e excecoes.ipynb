# Tratamento de erros


```python
print('O input abaixo pede um número inteiro ou não, pois está tipado. Sendo assim, podemos escrever qualquer coisa sem dar erro.\nEle também está dentro de um loop que vai receber os valores até que o user digite "0" para sair.')
print('---')

out = 1

while out == 1:
  print('Digite qualquer informação ou digite 0 para sair')
  a = input()
  print('Você digitou a informação:', a)
  print('---')

  if a == '0':
    out = 0
    break

print('---')
print('Agora vamos pedir especificamente um número inteiro. Porém, caso o user não digite isso, não dará erro.\nVamos tratar essa exceção com uma mensagem de saída.')
print('---')

while True:
  try:
    b = int(input('Digite um número inteiro: '))
  except:
    print('Você não digitou um valor inteiro')
    print('---')
  else:
    print('O número inteiro digitado foi:', b)
    break
```

```python
while True:
  try:
    a = int(input('Digite um número inteiro: '))
  except ValueError:
    print('O valor informado é inválido, não é um número inteiro.\nForneça um número inteiro...')
    print('---')
  except KeyboardInterrupt:
    print('O usuário interrompeu a execução')
    break
  else:
    print(f'Você digitou o número {a}')
    break
```
