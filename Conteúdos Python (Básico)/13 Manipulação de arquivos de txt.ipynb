# Manipulação de arquivos

- Importamos o arquivo txt pelo Files da barra lateral.
- O segundo parâmetro passado é o mode. De origem ele já vem como ler 'read (r)', mas podemos também escrever com o 'write (w)'

```python
 with open('/content/exemplo1.txt', 'r') as text:
  for linha in text:
    print(linha)
```

# Para manipular com python

```python
 with open('/content/exemplo1.txt') as text:
  r = text.readlines()
  print(r)
```

# Criar um arquivo novo

 -  Trocando o mode de read para write

```python
with open('/content/exemplo2.txt', 'w') as text:
  text.write('Essa linha foi escrita pelo Collab com os comandos write em python')
```

```python
with open('/content/exemplo2.txt') as text:
  print(text.read())
```
