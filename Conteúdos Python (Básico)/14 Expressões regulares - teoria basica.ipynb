# Expressões Regulares

- Busca sofisticada por padrões
- Procurar todas as palavras que começam com a letra maíscula, ou que começam com a letra maiúscula e terminam com a letra minúscula
- Que começam cmo 'a' e terminam com 'e'
- Procurar e-mails, número de telefone, CEP, etc.
- Padrões de busca são praticamente ilimitados

# Métodos

  - Search: Encontrar as posições de padrões dentro de uma string, se estes estiverem presentes.
  - Match: Encontrar se o começo e uma string é igual a um determinado padrão.
  - Findall: Encontrar todas as substrings em uma string que correspondam a um padrão.

# Metacaracteres:
- . - Qualquer caractere (exceto linha nova)
- \w - Qualquer caractere alfanumérico
- \W - Qualquer caractere não-alfanumérico
- \d - Qualquer caractere que seja um dígito(0-9)
- \D - Qualquer caractere não dígito
- \s - Espaço em branco
- ^ - Começa com
- $ - Termina com
- \ - Usado antes de metacaracteres para especificar seu significado literal

> OBS: Quer dizer que, se colocarmos '\\.' vamos procurar pelo caractere '.' específico e não por qualquer caractere.

# Quantificadores: Permitem determinar como e quantas vezes os metacaracteres aparecem.

- [] - Opcional entre os que estão dentro dos colchetes.
- () - Captura grupos de caracteres (0-9).
- '*' - de zero a mais vezes.
- ? - zero ou uma vez.
- '+' - uma ou mais vezes.
- {m,n} - de m a n vezes (Por exemplo, se tivermos um DDD que são 2 número colocaríamos (2,2).
- | - Ou

# Exemplos

- gabrielcosta@hotmail.com
- lazarento@gmail.com
- jenniferlopes@yahoo.com.br

> Representação com expressões regulares que detectam e-mails:

- \w+@\w+\.w+
- Isso quer dizer que:
 - Vamos pegar qualquer caractere alfanumérico (\w) uma ou mais vezes (+).
 - Vamos selecionar um caractere específico para pesquisa que é o '@' .
 - Vamos novamente pegar qualquer caractere alfanumérico (\w) uma ou mais vezes (+).
 - Pesquisar pelo caractere '.' específico com o (\.).
 - Mais uma sequência de caracteres alfanuméricos (\w) uma ou mais veze (+).
