# Função search

- Se não usarmos o r, talvez de algum tipo de aviso de sintaxe. O aviso é um lembrete para usar strings cruas (r"...") ao definir padrões de expressões regulares para evitar possíveis problemas de interpretação da barra invertida.

```python
import re # regular expression

frase1 = ('Ola, meu número de telefone é (81) 00000-0000')
frase2 = ('A placa de carro que eu anotei durante o acidente foi PFT8945')
frase3 = ('Entre em contato, meu email é t3ste@teste.com')

print(re.search(r'\(\d{2}\)\s\d{4,5}-\d{4}', frase1))
print(re.search(r'[A-Za-z]{3}\d{4}', frase2))
print(re.match(r'[A-Za-z]{3}\d{4}', frase2))
print(re.search(r'\w+@\w+\.com', frase3))
```

# Função Match

- Informa se a expressão regular está no início do texto

```python
frase2 = ('PFT8945 A placa de carro que eu anotei durante o acidente foi ')

print(re.match(r'[A-Za-z]{3}\d{4}', frase2))
```

#Findall

```python
frase1 = ('Ola, meu número de telefone é (81) 00000-0000 mas o meu número antigo é o (81) 11111-1111')

frase3 = ('Entre em contato, meu email é t3ste@teste.com ou tente no teste2@teste2.com')

print(re.findall(r'\(\d{2}\)\s\d{4,5}-\d{4}', frase1))
print(re.findall(r'\w+@\w+\.com', frase3))
```
