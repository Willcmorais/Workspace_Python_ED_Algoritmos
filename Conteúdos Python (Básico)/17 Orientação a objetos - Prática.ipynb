# Aqui vamos recriar a classe triangulo com suas funções.
- Que são o construtor (init), calcular a área e classificar o tipo.


```python
class Triangulo:

  # Método construtor com todos os atributos definidos
  def __init__(self, lado1, lado2, lado3, base, altura):
    self.lado1 = lado1
    self.lado2 = lado2
    self.lado3 = lado3
    self.base = base
    self.altura = altura

  def area(self):
    return (self.base * self.altura) / 2

  def tipo(self):
    if self.lado1 > self.lado2 + self.lado3 or self.lado2 > self.lado1 + self.lado3 or self.lado3 > self.lado1 + self.lado2:
      return 'não é um triângulo'
    elif self.lado1 == self.lado2 and self.lado2 == self.lado3:
      return 'um triângulo equilátero'
    elif self.lado1 == self.lado2 or self.lado1 == self.lado3 or self.lado2 == self.lado3:
      return 'um triângulo isósceles'
    elif self.lado1 != self.lado2 and self.lado1 != self.lado3 and self.lado2 != self.lado3:
      return 'um triângulo escaleno'

  def mostrar():
    print(f'Lados do triângulo: {t1.lado1}, {t1.lado2}, {t1.lado3} com base de {t1.base} e altura de {t1.altura}')
    print('---')
    print(f'A área calculada desse triângulo foi {Triangulo.area(t1)}')
    print('---')
    print(f'O triângulo foi classificado como {tipo1}')
```

```python
t1 = Triangulo(5, 6, 7, 8, 9)
tipo1 = Triangulo.tipo(t1)

Triangulo.mostrar()
```
