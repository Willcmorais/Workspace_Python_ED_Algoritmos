OPERADORES CONDICIONAIS:

- IF, ELSE e ELIF.

```python
num = int(input('Informe um número: '))

if (num > 7):
  print(f'O núm5ro {num} é maior que 7')
else:
  print(f'O número {num} não é maior que 7')
```

```python
x = 1
y = 5

if (x > 1) and (y % 2 == 0):
  print('x é maior que 1 e y é par')
else:
  print('Uma ou nenhuma das condições foram satisfeitas')

z = 1
w = 4

if (z > 1) or (w % 2 == 0):
  print('Ou x é maior que 1 ou y é par')
else:
  print('Uma ou nenhuma das condições foram satisfeitas')
```
