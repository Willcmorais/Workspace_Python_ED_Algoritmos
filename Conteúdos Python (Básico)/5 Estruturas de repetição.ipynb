- FOR

printa de 1 até 5. No ***range ()*** ele inicia do primeiro parâmetro informado e para antes do segundo parâmetro informado. Quando não acrescentamos o terceiro parâmetro que é a contagem, ele atribui o 1 como padrão.

```python
for num in range(1,6):
  print(num)
```

Aqui, como passamos o primeiro parâmetro para iniciar do 5, o segundo para parar antes do 0 e o terceiro para contar -1 a cada indice passado.

```python
for num in range(5,0,-1):
  print(num)
```

Por exemplo, vamos somar 5 + 4 + 3 + 2 + 1 e ver se também é igual a 15.

A soma se inicia em 0. Inicie o índice (i) no 1 e cubra um range até o número 6, não o incluindo. Após isso, pegue a soma e faça ela igual a ela mesma + o número do índice. Sendo assim, ele vai percorrer de 1 até o 5 e vai somar até o resultado final.

```python
soma = 0

for i in range (1,6):
  print(f'Somando {soma} com {i}...')
  soma = soma + i
  print(soma)
print(f'Resultado final: {soma}')
```

- Aqui nos vamos ter 3 índices i, j e k. Vamos iniciar com o índice i = 0 e parar a contagem. Depois vamos para o índice j = 0 e depois 1 e parar a contagem. E por último vamos para o k = 0, depois 1 e depois 2.

- Primeira vez do loop i agora ele vai ser igual a 1 e vai fazer tudo dnv, até chegar a 4.

```python
for i in range (0,5):
  print(i)
  print('contagem de i termina aqui')
  for j in range (0,2):
    print(j)
  print('contagem de j termina aqui')
  for k in range (0,3):
    print(k)
  print('contagem de k termina aqui')
```

- Aqui nós já vamos para outro conceito de fazer o i = 0 uma vez e parar. fazer o j = 0 uma vez e parar e depois fazer o k = 0, 1 e 2 e parar. Depois vamos voltar para o loop do j novamente igual a 1 e parando. sendo assim rodamos o loop do k novamente e paramos. finalizado o loop do j rodando o loop do k.

- Voltamos para o loop do i = 1 e fazemos tudo novamente.

```python
for i in range (0,5):
  print(i)
  print('contagem de i termina aqui')
  for j in range (0,2):
    print(j)
    print('contagem de j termina aqui')
    for k in range (0,3):
      print(k)
    print('contagem de k termina aqui')
```

- WHILE: nós precisamos criar a variável do índice para conseguir passar os parâmetros com o while. Após a contagem é necessário também que adiconemos manualmente um contador ao índice.

```python
i = 1

while i < 6:
  print(i)
  i += 1

print('---')

i = 5

while i > 0:
  print(i)
  i -= 1
```
