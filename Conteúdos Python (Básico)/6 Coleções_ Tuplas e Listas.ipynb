COLEÇÕES
- TUPLAS

As tuplas são representadas com ( ) e para atribuir um parâmetro com os índices utilizamos o [ ]. com a função ***index ( )*** ele printa o índice que está armazenado aquela info. Podemos também percorrer uma tupla com o comando FOR. O índice vai caminhar e printar todos os valores armazenados naquele espaço de memória.

```python
tupla = ('Homo sapiens', 'Cachorro', 'Gato')

print(tupla)
print()
print(tupla[1])
print()
print(tupla.index('Gato'))
print()

for i in tupla:
  print(i)
```
