1- Leia a idade do usuário e classifique-o em:

- Criança – 0 a 12 anos
- Adolescente – 13 a 17 anos
- Adulto – acima de 18 anos
-Se o usuário digitar um número negativo, mostrar a mensagem que a idade é inválida

```python
print('--------------------------------------------------------')
print('-------------------- CLASSIFICADOR ---------------------')
print('----------------------- ETÁRIO -------------------------')
print('--------------------------------------------------------\n')

#ler o nome
nome = input('Informe o seu nome: ').capitalize()
#ler a idade do usuário
idade = int(input('Informe a aqui a sua idade: '))
print('\nProcessando ....\n')

if idade > 0 and idade <= 12:
  print(f'{nome} tem {idade} anos e é uma criança')
elif idade >= 13 and idade <= 17:
   print(f'{nome} tem {idade} anos e é um(a) Adolescente')
elif idade >= 18:
   print(f'{nome} tem {idade} anos e é um(a) Adulto(a)')
elif idade <= -1:
  print('Idade informada inválida')

print('\n-------------------------------------------------------')
print('---------------------- OPERAÇÃO -----------------------')
print('--------------------- FINALIZADA ----------------------')
print('-------------------------------------------------------')
```

2- Calcular a média de um aluno que cursou a disciplina de Programação I, a partir da leitura das notas M1, M2 e M3; passando por um cálculo da média aritmética. Após a média calculada, devemos anunciar se o aluno foi aprovado, reprovado ou pegou exame.

- Se a média estiver entre 0.0 e 4.0, o aluno está reprovado
- Se a média estiver entre 4.1 e 6.0, o aluno pegou exame
- Se a média for maior do que 6.0, o aluno está aprovado
- Se o aluno pegou exame, deve ser lida a nota do exame. Se a nota do exame for maior do que 6.0, está aprovado, senão; está reprovado

```python
print('*************************************************************')
print('************************ BEM VINDO À ************************')
print('*********************** DISCIPLINA DE ***********************')
print('*********************** PROGRAMAÇÃO I ***********************')
print('*************************************************************')

print('\nVamos calcular as médias da disciplina e avaliar, se o aluno será aprovado, reprovado ou ficará de recuperação.\n')

nome = input('Informe seu nome: ').capitalize()
print('\nProcessando ....\n')
nota_01 = float(input('Informe a primeira nota: '))
nota_02 = float(input('Informe a segunda nota: '))
nota_03 = float(input('Informe a terceira nota: '))
print('\nProcessando ....\n')

#arredonda a média com o round
media_notas = round((nota_01 + nota_02 + nota_03) / 3, 2)

print(f'A média do aluno {nome} foi de {media_notas}\n')

if media_notas >= 0 and media_notas <=4:
  print('O aluno {} está reprovado'.format(nome))
elif media_notas > 4 and media_notas < 7:
  print('O aluno {} poderá ir para à recuperação'.format(nome))
elif media_notas >= 7:
  print('O aluno {} está aprovado'.format(nome))
else:
  print('Nota inválida')

print('\n-------------------------------------------------------')
print('---------------------- OPERAÇÃO -----------------------')
print('--------------------- FINALIZADA ----------------------')
print('-------------------------------------------------------')
```
