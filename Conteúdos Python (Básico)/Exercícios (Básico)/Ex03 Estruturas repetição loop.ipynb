1- Ler 5 notas e informar a média

```python
#USANDO WHILE
i = 0
j = 1
soma = 0

print('EXEMPLO USANDO O WHILE\n')

while i < 1:
  nota = float(input('informe a sua nota: '))
  soma = soma + nota
  print('A soma atual das notas é',(soma),'\n')
  i += 1

while j < 5:
  nota = float(input('informa a próxima nota: '))
  soma = soma + nota
  print('A soma atual das notas é',(soma),'\n')
  j += 1

media = soma / 5

print('O print da média no format WHILE foi de: {}'.format(round(media, 2)),'\n')

#USANDO FOR
soma = 0

print('EXEMPLO USANDO O FOR\n')

for i in range (1):
  nota = float(input('Informe a sua nota: '))
  soma = soma + nota
  print('A soma atual das notas é',(soma),'\n')

for j in range (1,5):
  nota = float(input('Informe a próxima nota: '))
  soma = soma + nota
  print('A soma atual das notas é',(soma),'\n')

media = soma / 5

print('O print da média no format FOR foi de: {}'.format(round(media, 2)),'\n9')
```

2- Imprimir a tabuada do número 3 (3 x 1 = 1 - 3 x 10 = 30)

```python
i = 1
num = 3

while i < 11:
  result = num * i
  print(f'{num} x {i} = {result}')
  i += 1
```
