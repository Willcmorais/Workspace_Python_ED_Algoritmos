2.   Dicionário: Crie um dicionário para armazenar o nome e a nota de 3 alunos, fazendo a leitura dos valores por meio de uma estrutura de repetição. Depois, crie uma nova estrutura de repetição para somar todas as notas e retornar a média.



```
alunos = {}
for _ in range(1, 4):
  nome = input('Digite o nome: ')
  nota = float(input('Digite a nota: '))
  alunos[nome] = nota

soma = 0
for nota in alunos.values():
  soma += nota
print('Média: ', soma / 3)
```

````python
print('=====================================================')
print('=================== ARMZ DE NOMES,===================')
print('=================== SOMA DE NOTAS ===================')
print('==================== E MÉDIAS C/=====================')
print('==================== DICIONÁRIO =====================')
print('=====================================================\n')

dicionario = {}
tamanho = 3
soma = 0
i = 0

while i < tamanho:
  nome = input('Digite o nome do aluno: ')
  nota = float(input('Digite a nota do aluno: '))
  print('---')
  dicionario[nome] = nota
  soma = soma + nota
  i += 1

for nome, nota in dicionario.items():
    print(f'Nome: {nome}\nNota: {nota}')
    print('---')

print('A média de notas dos 3 alunos foi de {}'.format(round(soma/tamanho, 1)))
print('---')
print('Fim da operação....')
````

3.   Matriz: Dada a matriz abaixo, construa uma estrutura de repetição para percorrer e somar todos os elementos da matriz.



```
matriz = np.array([[3, 4, 1],
                   [3, 1, 5]])

soma = 0
for i in range(matriz.shape[0]):
  for j in range(matriz.shape[1]):
    soma += matriz[i][j]
print('Soma: ', soma)
```



````python
import numpy as np

matriz = np.array([[3, 4, 1],[3, 1, 5]])

print('Matriz criada:\n',matriz)
print('---')

soma = 0

for i in range(matriz.shape[0]):
  for j in range(matriz.shape[1]):
    soma = soma + matriz[i][j]

print(f'A soma dos valores dessa matriz é {soma}')
print('---')
print('Fim da operação....')
````
