# Crie expressões regulares para extrair as seguintes informações do texto abaixo (use a função findall):
- Números
- CEPs
- URLs

```python
import re

texto = ('Me chamo José alberto carneiro pinto. Tenho 34 anos, nasci em teresina-PI. Moro na rua joão goulart neto filho genro, n 3335, apto 1102, CEP 55160-910. O site que mais frequento hoje em dia é o https://www.youtube.com.br e também gosto de assistir também o http://www.disneyplus.com')

print('Lista de todos os números que foram encontrados juntos:\n',re.findall(r'\d+', texto))
print('---')
print('Lista de todos os números que foram encontrados:\n', re.findall(r'\d', texto))
print('---')
print('Lista de todos os números em formato de CEPs encontrados:\n',re.findall(r'\d{5}-\d{3}', texto))
print('---')
print('Lista de todos os caracteres em formatos de URLs encontrados:\n',re.findall(r'https?://[A-Za-z0-9./]+', texto))
```
