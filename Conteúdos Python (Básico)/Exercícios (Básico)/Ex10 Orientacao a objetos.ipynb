# Crie uma classe chamada aluno com os seguintes atributos:
- Nome
- Nota 1
- Nota 2
- Crie um construtor para a classe (__init__)

# Crie as seguintes funções (métodos):
- Calcula média, retornando a média aritmética entre as notas
- Mostra dados, que somente imprime o valor de todos os atributos
- Resultado, que verifica se o aluno está aprovado ou reprovado (se a média for maior ou igual a 6.0, o aluno está aprovado)

# Crie três objetos (aluno1, aluno2 e aluno3) e teste as funções

```python
class Aluno:

  def __init__(self, nome, nota1, nota2):
    self.nome = nome
    self.nota1 = nota1
    self.nota2 = nota2

  def calcular_media(self):
    return (self.nota1 + self.nota2) / 2

  def avaliar_media(self):
    if (Aluno.calcular_media(self)) < 5:
      print('O aluno(a) está reprovado')
    elif (Aluno.calcular_media(self) < 7):
      print('O aluno(a) está na recuperação')
    else:
      print('O aluno(a) está aprovado!')

  def mostrar_dados(self):
    return print(f'O aluno(a) {self.nome} tirou {self.nota1} na AV1 e {self.nota2} na Av2. Sendo assim, sua média foi de {self.calcular_media()}.')
```

```python
aluno1 = Aluno('William coelho de morais', 5.5, 4)
aluno1.mostrar_dados()
aluno1.calcular_media()
aluno1.avaliar_media()
print('---')
aluno2 = Aluno('Victoria Maria Ferreira dos Anjos', 10, 8.5)
aluno2.mostrar_dados()
aluno2.calcular_media()
aluno2.avaliar_media()
print('---')
aluno3 = Aluno('Gabriel Vidal de Moura', 6.5, 5.5)
aluno3.mostrar_dados()
aluno3.calcular_media()
aluno3.avaliar_media()
```
