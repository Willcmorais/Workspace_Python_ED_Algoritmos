Crie uma função que determine se um número é par ou ímpar e, adicionalmente, se ele é positivo, negativo ou nulo.

```python
num = int(input('Digite um número: '))

if num % 2 == 0:
  print(f'O número {num} é par')
else:
  print(f'O número {num} é impar')

if num > 0:
  print(f'O número informado {num} é um número positivo')
elif num < 0:
  print(f'O número informado {num} é um número negativo')
else:
  print(f'O número informado {num} é um número nulo')
```

Implemente uma função que receba um número inteiro n e retorne uma lista com os n primeiros números primos.

- num = 2  # primeiro número candidato

- num ** 0.5

Eleva num a 0.5, o que é o mesmo que tirar a raiz quadrada.

Por exemplo: 9 ** 0.5 → 3.0.

Por que só até a raiz quadrada?

Se num não tiver nenhum divisor menor ou igual à sua raiz quadrada, ele não terá divisores maiores (exceto ele mesmo).

Exemplo: se num = 36, a raiz é 6.
Se não acharmos divisor até 6, significa que não haverá nenhum divisor depois disso (além do próprio 36).

Isso reduz muito a quantidade de testes, deixando o código mais rápido.

O + 1

O range() em Python vai até o valor especificado, mas não o inclui.

Como queremos incluir o valor da raiz quadrada no teste, somamos + 1.

```python
print('Vou te mostrar uma lista de números primos.')

n = int(input("Digite a quantidade de números primos desejada: "))

primos = []
# Inicializa a variável num com o valor 2. num representa o número atual que está sendo verificado para saber se é primo. Começamos com 2, pois é o primeiro número primo.
num = 2

while len(primos) < n:
  for i in range(2, int(num ** 0.5) + 1): # vazio
      if num % i == 0:
          break
  else:
      primos.append(num)
  num += 1

print(primos)
```
