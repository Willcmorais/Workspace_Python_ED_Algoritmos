- MANIPULAÇÃO DE STRINGS:

Criando variáveis e manipulando com as funções ***upper, lower, capitalize, title e chamando o índice de cada letra com [n:n]***.

```python
a = 'casaco de couro'
maiuscula = a.upper()
minuscular = a.lower()
capitalizar = a.capitalize()
title = a.title()
metade_palavra = a[0:9]
ultimas_letras = a[7:]

print(maiuscula)
print(minuscular)
print(capitalizar)
print(title)
print(metade_palavra)
print(ultimas_letras)
```

- Utilizando a função ***replace( )*** para alterar um campo da variável

```python
b = a.replace('aco','inha')
print(a)
print(b)
```

- Criamos aqui uma variável C que recebe o valor de A com os caracteres 'a' ao invés de 'o' na palavra. Depois pegamos essa palavra e procuramos com a função ***find ()***, qual é a posição índice em que a primeira letra 's' aparece na palavra.

```python
c = a.replace('o','a')
print(c)

letra = input('Informe aqui a letra que deseja procurar: ')

if(c.find(letra) != -1):
  print('A primeira posição que a letra {} apareceu foi na {}'.format(letra, c.find(letra)))
elif(c.find(letra) == -1):
  print(f'A primeira posição de {letra} não pode ser informada, pois, não existe esta letra na palavra.')
```

- Com a função ***len ()*** podemos descobrir o tamanho de caracteres da string. Caso tenha espaços em branco no início ou no final da palavra, podemos cortar com a função ***strip ()***.

```python
e = ' casaco de couro '
print(len(a))
print(len(e))

f = e.strip()
print(len(f))
```
