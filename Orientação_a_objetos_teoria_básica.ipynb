# Orientação a objetos - Classe e objeto

- Agrupamento de objetos similares que representam os mesmos atributos (características/variáveis) e métodos (funções).

- Podemos analisar como se fosse um molde de bonecos de gesso:
 - Definimos o formato e tamanho.
 - O molde é sempre o mesmo, porém, os objetos gerados podem ter características variadas, respeitando a estrutura básica do molde.

- A classe, em orientação a objetos, seria o molde e o objeto seriam as possíveis variáveis criadas com este molde.

> # Por exemplo:
- Podemos ter uma classe Cachorro e com ela podemos criar vários outros objetos com características (atributos) diferentes, mas continuam sendo cachorros. Por exemplo um beagle, labrador, dalmata.
- Todos eles tem características (atributos) em comum. Como cor, tamanho, número de patas e raça.
- Todos eles também fazem ações (métodos) semelhantes. Eles podem, por exemplo, latir, correr, morder, brincar e dormir.

> # Outro exemplo:
- Uma calculadora.
- Características (atributos/variáveis): cor, tamanho, marca, data de fabricação.
- Ações (métodos/funções): liga, desliga, soma, subtrai, divide e multiplica.

> OBS: UMA CLASSE TEM CARACTERÍSTICAS (ATRIBUTOS/VARIÁVEIS) E MÉTODOS (AÇÕES/FUNÇÕES)

- Levando para o código, características ou atributos podem ser strings, listas, tuplas, float, etc.
- Os métodos são as ações ou funções que aquela classe oferece para criação de objetos.
