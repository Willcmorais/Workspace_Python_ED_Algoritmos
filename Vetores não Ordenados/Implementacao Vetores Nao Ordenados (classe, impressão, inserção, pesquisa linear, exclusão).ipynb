> Criação da classe do vetor e as suas funções
- Método init (construtor): Os parâmetros que já vão existir e os que serão atribuidos.

- Função de inserção de valores (insere).

- Função de pesquisa linear de valores (pesquisa).

- Função de pesquisa, exclusão e realocação de valores do vetor.

- Função que imprime as informações (imprime).


```python
class VetorNaoOrdenado():

  def __init__(self, capacidade):
    '''
    - ultima_posicao: Ponteiro que vai apontar aonde está o último valor do vetor.
    - valores: Recebe uma lista vazia de números inteiros do tamanho da capacidae atribuida.
    '''
    self.capacidade = capacidade
    self.ultima_posicao = -1
    self.valores = np.empty(self.capacidade, dtype=int)

  def insere(self, valor):
    '''
    - Primeiro if: código de segurança. Para não estrapolar o tamanho máximo do vetor.
    - else: Vai adicionar 1 para o índice de última posição e adicionar o valor fornecido no índice da última posição da lista.
    '''
    if self.ultima_posicao == self.capacidade -1:
      print('Capacidade máxima atingida.')
    else:
      self.ultima_posicao += 1
      self.valores[self.ultima_posicao] = valor

  def pesquisa(self, valor):
    '''
    - Primeiro if: Código de segurança quando a lista está vazia.
    - else + for: Enquanto estiver entre i = 0 e o valor da última posição + 1 (fazemos isso para que todos os valores sejam lidos).
    - Segundo if: Se o valor que estiver no índice e o valor fornecido forem iguais, ele vai retornar o índice. Se não, vai retornar - 1. Indicando que não existe.
    '''
    if self.ultima_posicao == -1:
      print('Pesquisa indisponível, o vetor está vazio.')
    else:
      for i in range(self.ultima_posicao + 1):
        if valor == self.valores[i]:
          return i
      return -1

  def exclui(self, valor):
    '''
    - posicao: A variável vai receber um valor que vai ser utilizado pela função (pesquisa) dentro da função (exclui). Isso é feito para achar se o valor informado está no vetor para assim ser excluído.
    - Primeiro if: Se a posicao for igual a - 1 isso quer dizer que o número não existe na lista. Então, não pode ser excluído (tem return).
    - else + for: Enquanto estiver entre o range do valor da posicao[i] e valor[i] da última posição, ele vai pegar o valor do i aual + 1 e jogar no i atual. Fazemos isso para que todos os valores sejam realocados a partir do valor excluído. Depois, é subtraido 1 da ultima_posicao para realocar o apontador junto com os valores também.
    '''
    posicao = self.pesquisa(valor)
    if posicao == -1:
      return -1
    else:
      for i in range(posicao, self.ultima_posicao):
        self.valores[i] = self.valores[i + 1]
      self.ultima_posicao -= 1

  def imprime(self):
    '''
    - Primeiro if: Se a última posição for - 1 quer dizer que está vazio.
    - else + for: Printa todos os valores dos índices no range predefinido.
    '''
    if self.ultima_posicao == -1:
      print('O vetor está vazio.')
    else:
      for i in range(self.ultima_posicao + 1):
        print(f'Índice {i}:\nValor - {self.valores[i]}')
        print('---')
```
